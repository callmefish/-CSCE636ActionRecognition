```python
import os
import csv
import matplotlib.pyplot as plt


file = 'record/spatial_497_sim_3/rgb_test.csv'
test_loss_3 = []
test_acc_3 = []
with open(file, 'r') as f:
    reader = csv.reader(f)
    # 一行一行读取CSV文件，记录数值
    i = 0
    for row in reader:
        if i < 1:
            i += 1
            continue
        else:
            test_loss_3.append(round(float(row[2]), 6))
            test_acc_3.append(round(float(row[3])*0.01, 6))
f.close()
```

```python
file = 'record/spatial_497_sim_3/rgb_train.csv'
train_loss_3 = []
train_acc_3 = []
with open(file, 'r') as f:
    reader = csv.reader(f)
    # 一行一行读取CSV文件，记录数值
    i = 0
    for row in reader:
        if i < 1:
            i += 1
            continue
        else:
            train_loss_3.append(round(float(row[3]), 6))
            train_acc_3.append(round(float(row[4])*0.01, 6))
f.close()
```

```python
plt.plot(train_loss_3, label='Training loss')
plt.plot(test_loss_3, label='Validation loss')
# plt.ylim(0, int(max(Loss)) + 1)
plt.title("TSN - 3 frame - Loss - 8")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.ylim(0, 10)
plt.legend()
#plt.savefig("record/TSN - 3 frame - Loss.png")
plt.show()
```

```python
plt.plot(train_acc_3, label='Training acc')
plt.plot(test_acc_3, label='Validation acc')
# plt.ylim(0, int(max(Loss)) + 1)
plt.title("TSN - 3 frame - Acc - 8")
plt.xlabel("Epoch")
plt.ylabel("Acc")
plt.legend()
#plt.savefig("record/TSN - 3 frame - Acc.png")
plt.show()
```
